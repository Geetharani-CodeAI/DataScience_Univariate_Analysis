```python
# Replacing all nulls with mean value using Imputer
```

```python
df1.dtypes
```

```python
def qualQuan(df1):
    quan=[]
    qual=[]
    for columnName in df1.columns:
        if df1[columnName].dtypes =='O':
            qual.append(columnName)
        else:
            quan.append(columnName)
    return qual,quan
        
 
```

```python
qual,quan = qualQuan(df1)
```

```python
qual
```

```python
quan
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imp.fit_transform(df1[quan])
df2 = pd.DataFrame(df1[quan])
```

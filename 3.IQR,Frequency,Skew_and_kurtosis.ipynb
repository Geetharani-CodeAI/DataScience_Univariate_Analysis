```python
def quanQual(dataset):
    quan = []
    qual = []
    for columnName in dataset.columns:
        if dataset[columnName].dtypes == 'O':
            qual.append(columnName)
        else:
            quan.append(columnName)    
    return qual,quan
        
```

Inter Quartile Range (IQR)
----------------------------

```python

descriptive = pd.DataFrame(index = ["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","Q4:100%","IQR","1.5 Rule","Lesser","Greater","Min"],columns=quan)
for columnName in quan:
    descriptive[columnName]["Mean"] = dataset[columnName].mean().astype(int)
    descriptive[columnName]["Median"] = dataset[columnName].median()
    descriptive[columnName]["Mode"] = dataset[columnName].mode()[0]
    descriptive[columnName]["Q1:25%"] = dataset.describe()[columnName]["25%"]
    descriptive[columnName]["Q2:50%"] = dataset.describe()[columnName]["50%"]
    descriptive[columnName]["Q3:75%"] = dataset.describe()[columnName]["75%"]
    descriptive[columnName]["Q4:100%"] = dataset.describe()[columnName]["max"]
    descriptive[columnName]["IQR"] = descriptive[columnName]["Q3:75%"] - descriptive[columnName]["Q1:25%"]
    descriptive[columnName]["1.5 Rule"] = 1.5 * descriptive[columnName]["IQR"]
    descriptive[columnName]["Lesser"] = descriptive[columnName]["Q1:25%"] - descriptive[columnName]["1.5 Rule"] 
    descriptive.loc["Greater",columnName] = descriptive[columnName]["Q3:75%"] + descriptive[columnName]["1.5 Rule"] 
    descriptive.loc["Min", columnName] = dataset[columnName].min()
    descriptive.loc["Max",columnName] = dataset[columnName].max()
```

Finding Outliers
-----------------------

```python
def findOutlier(columnName,quan):
    lesser = []
    greater = []
    for columnName in quan:
        if (descriptive[columnName]["Min"] < descriptive[columnName]["Lesser"]):
            lesser.append(columnName)
        if (descriptive[columnName]["Max"] > descriptive[columnName]["Greater"]):
            greater.append(columnName)
    return lesser,greater
```

```python
lesser,greater = findOutlier(columnName,quan)
```

```python
def ReplaceOutlier(columnName):
    for columnName in lesser:
        dataset[columnName][dataset[columnName] < descriptive[columnName]["Lesser"]] = descriptive[columnName]["Lesser"]
    for columnName in greater:
        dataset[columnName][dataset[columnName] > descriptive[columnName]["Greater"]] = descriptive[columnName]["Greater"]
    return lesser,greater
```

```python
def univariate(dataset,quan):
    descriptive = pd.DataFrame(index = ["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","Q4:100%","IQR","1.5 Rule",
                                        "Lesser","Greater","Min","Max","Skewness","Kurtosis"],columns=quan)
    for columnName in quan:
        descriptive[columnName]["Mean"] = dataset[columnName].mean().astype(int)
        descriptive[columnName]["Median"] = dataset[columnName].median()
        descriptive[columnName]["Mode"] = dataset[columnName].mode()[0]
        descriptive[columnName]["Q1:25%"] = dataset.describe()[columnName]["25%"]
        descriptive[columnName]["Q2:50%"] = dataset.describe()[columnName]["50%"]
        descriptive[columnName]["Q3:75%"] = dataset.describe()[columnName]["75%"]
        descriptive[columnName]["Q4:100%"] = dataset.describe()[columnName]["max"]
        descriptive[columnName]["IQR"] = descriptive[columnName]["Q3:75%"] - descriptive[columnName]["Q1:25%"]
        descriptive[columnName]["1.5 Rule"] = 1.5 * descriptive[columnName]["IQR"]
        descriptive[columnName]["Lesser"] = descriptive[columnName]["Q1:25%"] - descriptive[columnName]["1.5 Rule"] 
        descriptive.loc["Greater",columnName] = descriptive[columnName]["Q3:75%"] + descriptive[columnName]["1.5 Rule"] 
        descriptive.loc["Min", columnName] = dataset[columnName].min()
        descriptive.loc["Max",columnName] = dataset[columnName].max()  
        descriptive[columnName]["Skewness"] = dataset[columnName].skew()
        descriptive[columnName]["Kurtosis"] = dataset[columnName].kurtosis()
    return descriptive
```

```python
descriptive = univariate(dataset,quan)
```
